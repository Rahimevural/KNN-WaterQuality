```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("waterQuality1.csv")
data = data.replace('#NUM!', pd.NA)
data = data.dropna()
data.head()
#data.tail()
```

```python
data1=data.drop(["is_safe"],axis=1)
data1.head(5)
```

```python
from re import X
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data1,data.is_safe,test_size=0.4, random_state=0)

```

```python
print(x_train)
```

```python
print(y_test)
```

```python
from sklearn.preprocessing import LabelEncoder
from collections import Counter
categorical_cols = x_train.select_dtypes(include=['object']).columns
encoders = {}
for col in categorical_cols:
    encoders[col] = LabelEncoder()
    # Fit on training data
    encoders[col].fit(x_train[col])

    # Get most frequent value for imputation
    most_frequent_value = Counter(x_train[col]).most_common(1)[0][0]

    # Transform training data
    x_train[col] = encoders[col].transform(x_train[col])

    # Transform test data, replacing unknown values with the most frequent value
    x_test[col] = x_test[col].map(lambda s: s if s in encoders[col].classes_ else most_frequent_value)
    x_test[col] = encoders[col].transform(x_test[col])
```

```python
#model
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=11)
knn.fit(x_train,y_train)
tahmin = knn.predict(x_test)
```

```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

score_list = []
for each in range(1, 15):  # k değerlerini 1'den 14'e kadar test ediyoruz
    knn1 = KNeighborsClassifier(n_neighbors=each)
    knn1.fit(x_train, y_train)  # Modeli eğitim verisiyle eğitiyoruz
    score_list.append(knn1.score(x_test, y_test))  # Test skoru hesaplanıp listeye ekleniyor

# Grafik çizimi döngü tamamlandıktan sonra yapılır
plt.title("En uygun k değerine göre score")
plt.plot(range(1, 15), score_list, color="red")
plt.xlabel("k değeri")
plt.ylabel("score")
plt.show()
```
